```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate the Euclidean distance between two points
def distance(city1, city2):
    return math.sqrt((city1[0] - city2[0])**2 + (city1[1] - city2[1])**2)

# Calculate the total distance of the path (tour)
def total_distance(path, cities):
    dist = 0
    for i in range(len(path) - 1):
        dist += distance(cities[path[i]], cities[path[i + 1]])
    dist += distance(cities[path[-1]], cities[path[0]])  # Return to the start city
    return dist

# Swap two cities in the tour
def swap(path):
    a, b = random.sample(range(len(path)), 2)
    path[a], path[b] = path[b], path[a]

# Simulated Annealing algorithm
def simulated_annealing(cities, initial_temperature, cooling_rate, max_iter):
    # Create an initial random path (tour)
    path = list(range(len(cities)))
    random.shuffle(path)
    
    # Calculate the initial distance
    current_distance = total_distance(path, cities)
    best_path = path[:]
    best_distance = current_distance
    
    temperature = initial_temperature
    
    # Main Simulated Annealing loop
    for i in range(max_iter):
        # Make a small change to the current path
        new_path = path[:]
        swap(new_path)
        
        # Calculate the new distance
        new_distance = total_distance(new_path, cities)
        
        # Accept the new path if it's better or based on a probability
        if new_distance < current_distance or random.random() < math.exp((current_distance - new_distance) / temperature):
            path = new_path
            current_distance = new_distance
            
            # Update the best solution
            if current_distance < best_distance:
                best_path = path[:]
                best_distance = current_distance
        
        # Cool the temperature
        temperature *= cooling_rate
    
    return best_path, best_distance

# Plot the cities and the best path
def plot_cities(cities, best_path):
    x = [cities[i][0] for i in best_path] + [cities[best_path[0]][0]]
    y = [cities[i][1] for i in best_path] + [cities[best_path[0]][1]]
    plt.plot(x, y, marker='o')
    plt.scatter([city[0] for city in cities], [city[1] for city in cities], color='red')
    plt.title('Best Path Found by Simulated Annealing')
    plt.show()

# Example Usage:
# Define a set of cities (coordinates)
cities = [(random.randint(0, 100), random.randint(0, 100)) for _ in range(10)]

# Simulated Annealing parameters
initial_temperature = 10000
cooling_rate = 0.995
max_iter = 10000

# Solve TSP using simulated annealing
best_path, best_distance = simulated_annealing(cities, initial_temperature, cooling_rate, max_iter)

print(f"Best Path: {best_path}")
print(f"Best Distance: {best_distance}")

# Plot the cities and the best path
plot_cities(cities, best_path)
```
